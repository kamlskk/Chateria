````python
import pprint

# Zmodyfikowane fakty
fakty = {
    'ojciec': [
        ('Jan', 'Anna'),
        ('Piotr', 'Kasia'),
        ('Józef', 'Jan'),
        ('Edward', 'Piotr'),
        ('Stanisław', 'Józef'),
        ('Henryk', 'Edward'),
        ('Feliks', 'Stanisław'),
        ('Władysław', 'Henryk')
    ],
    'matka': [
        ('Maria', 'Anna'),
        ('Ewa', 'Kasia'),
        ('Barbara', 'Maria'),
        ('Ewa', 'Piotr'),
        ('Zofia', 'Barbara'),
        ('Helena', 'Ewa'),
        ('Karolina', 'Zofia'),
        ('Maria', 'Helena')
    ]
}

# Funkcja do sprawdzania, czy osoba1 jest matką osoby2
def czy_jest_matką(fakty, matka, dziecko):
    return (matka, dziecko) in fakty.get('matka', [])

# Funkcja do sprawdzania, czy osoba1 jest rodzicem osoby2
def czy_jest_rodzicem(fakty, rodzic, dziecko):
    return (rodzic, dziecko) in fakty.get('ojciec', []) or (rodzic, dziecko) in fakty.get('matka', [])

# Funkcja do dodawania faktów do tabeli

def dodaj_fakt(fakty, rodzaj, osoba1, osoba2):
    if rodzaj not in fakty:
        fakty[rodzaj] = []
    fakty[rodzaj].append((osoba1, osoba2))

# Funkcja do sprawdzania dowolnej relacji
def sprawdz_relacje(relacje, rodzaj, osoba1, osoba2):
    return (osoba1, osoba2) in relacje.get(rodzaj, [])

# Funkcja do generowania relacji dziadków, pradziadków i pra-pra-dziadków
def generuj_relacje(fakty):
    relacje = fakty.copy()
    relacje['dziadek'] = []
    relacje['babcia'] = []
    relacje['pradziadek'] = []
    relacje['prababcia'] = []
    relacje['prapradziadek'] = []
    relacje['praprababcia'] = []

    # Dodanie dziadków
    for rodzaj in ['ojciec', 'matka']:
        for rodzic, dziecko in fakty[rodzaj]:
            for wnuk in [wnuk for rodzic2, wnuk in fakty['ojciec'] + fakty['matka'] if rodzic2 == dziecko]:
                if rodzaj == 'ojciec':
                    relacje['dziadek'].append((rodzic, wnuk))
                elif rodzaj == 'matka':
                    relacje['babcia'].append((rodzic, wnuk))

    # Dodanie pradziadków
    for rodzaj in ['dziadek', 'babcia']:
        for dziadek_babcia, wnuk in relacje[rodzaj]:
            for prawnuk in [prawnuk for rodzic, prawnuk in fakty['ojciec'] + fakty['matka'] if rodzic == wnuk]:
                if rodzaj == 'dziadek':
                    relacje['pradziadek'].append((dziadek_babcia, prawnuk))
                elif rodzaj == 'babcia':
                    relacje['prababcia'].append((dziadek_babcia, prawnuk))

    # Dodanie pra-pra-dziadków
    for rodzaj in ['pradziadek', 'prababcia']:
        for pradziadek_babcia, prawnuk in relacje[rodzaj]:
            for praprawnuk in [praprawnuk for rodzic, praprawnuk in fakty['ojciec'] + fakty['matka'] if rodzic == prawnuk]:
                if rodzaj == 'pradziadek':
                    relacje['prapradziadek'].append((pradziadek_babcia, praprawnuk))
                elif rodzaj == 'prababcia':
                    relacje['praprababcia'].append((pradziadek_babcia, praprawnuk))

    return relacje

# Generowanie wszystkich relacji (w tym dziadków i pradziadków)
relacje = generuj_relacje(fakty)

# Przykładowe zapytania
print("Czy Edward jest ojcem Piotra?", czy_jest_rodzicem(fakty, 'Edward', 'Piotr'))
print("Czy Maria jest matką Anny?", czy_jest_matką(fakty, 'Maria', 'Anna'))
print("Czy Stanisław jest dziadkiem Piotra?", sprawdz_relacje(relacje, 'dziadek', 'Stanisław', 'Piotr'))
print("Czy Feliks jest pradziadkiem Kasi?", sprawdz_relacje(relacje, 'pradziadek', 'Feliks', 'Kasia'))
print("Czy Władysław jest prapradziadkiem Piotra?", sprawdz_relacje(relacje, 'prapradziadek', 'Władysław', 'Piotr'))
````
